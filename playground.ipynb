```python
textrank_outputs = [keywords.keywords(doc).split('\n') for doc in doc_complete]
rake_outputs = []
for doc in doc_complete:
    r = Rake(max_length=2) # Uses stopwords for english from NLTK, and all puntuation characters.
    r.extract_keywords_from_text(doc)
    rake_outputs.append(r.get_ranked_phrases())
```
